```python
import numpy as np
import matplotlib.pyplot as plt
from numpy.random import uniform as ran

samples = 1000
trials = 1000
tmin = 82
tmax = 94
frequencies = np.linspace(tmin,tmax,tmax-tmin+1)
pi2 = 2*np.pi
```

```python
wf = 10
samples = 1000
samps = np.linspace(0,1,num=samples+1)
samp_rate = wf*len(samps)
waveTable = [np.cos(s*pi2) for s in samps]

def genWave(freq,phase):
    p = phase/pi2*1000
    wave = [waveTable[int((s*freq/wf+p)%len(samps))] for s in range(len(samps))]
    return wave,phase
```

```python
'''
Random Phase
'''
# plt.figure(figsize=(15,10))

# plt.ylim(bottom=-32,top=32)
x = samps
sinewaves = [genWave(y,0) for y in range(tmin,tmax)]
sin = np.sum(sinewaves,axis=0)

plt.plot(x,sin)
plt.show()

maxvals = []
for i in range(trials):
    plt.ylim(bottom=-32,top=32)
    
    sinewaves = [genWave(w,ran(0,pi2)) for w in range(tmin,tmax)]
    sin = np.sum(sinewaves,axis=0)

    maxvals.append(np.amax(sin))
        
    plt.plot(x,sin)
    plt.show()
    print(maxvals[-1])
plt.plot(np.linspace(1,trials,num=trials),maxvals)

plt.show()
```

```python
'''
Minimization
'''
plt.figure(figsize=(15,10))
x = samps


wave = []
phases = [0]
for w in frequencies:
    #Save first cosine wave
    if w==tmin:
        wave = (genWave(w,0))[0]
    else:
        tempwaves = []
        maxima = []
        #Add next wave in series, incrementing the phase
        for i in range(samples):
            tempcos = (genWave(w,i*pi2))[0]
            tempwaves.append(np.add(tempcos,wave))
            maxima.append(np.amax(tempwaves[-1]))
#         plt.plot(x,wave)
#         plt.show()
        pind = np.argmin(maxima)
        wave=tempwaves[pind]
        plt.plot(x,(genWave(w,pind*pi2)[0]))
        phases.append(pind)
        print(w,pind*pi2)


#     plt.plot(x,wave)
#     plt.show()
plt.show()
plt.figure(figsize=(15,10))
plt.plot(x,wave)
plt.show()
print(phases)            
            
   
```
